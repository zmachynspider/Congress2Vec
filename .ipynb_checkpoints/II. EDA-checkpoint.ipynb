## EDA

```python
#load in the data
df = pd.read_pickle('data/gov_doc_df.pkl')
```

```python
df.head(2)
```

```python
speaker_df = df[(df['party'] != 'unavailable') & (df['protocol'] == False)].groupby(['speaker','party','state'])['length'].apply(lambda x: x.sum()).to_frame()
speaker_df = pd.melt(speaker_df.reset_index(), id_vars=['speaker','party','state'])
```

```python
#important to exclude protocol speech to see notice these trends
df[(df['party'] != 'unavailable') & (df['protocol'] == False)].hist(column='length', by='party', bins=50, normed = 1);
```

```python
speaker_df.hist(column='value', by='party', bins=50, normed = 1);
```

> *Although it's not as clear on a per-record basis, when we look at word count by representative, it's obvious that the Democratic party is the party of words. Perhaps it is unremarkable that word count for both parties is right-skewed. Although Democrats have more words, let's explore who is responsible for the long-tail in the Republican party*

```python
#Who are the bit outliers
speaker_df[(speaker_df.value-speaker_df.value.mean())>=(3*speaker_df.value.std())] #keep only the ones that are within +3 to -3 standard deviations in the column 'Data'.
```

## Words by State by Rep

*California, Texas, and NY have the most reps in our dataset*

```python
ordered_state = speaker_df['state'].value_counts().index
plt.rcParams['figure.figsize']=(11,5)
p = sns.countplot(x='state', data=speaker_df, order=ordered_state)
p.set_xticklabels(p.get_xticklabels(), rotation=90)
p.set(xlabel='State', ylabel='Representatives')
sns.plt.title('Reps by State');
```

```python
state_df = df[df['protocol'] == False].groupby('state')['length'].apply(lambda x: x.sum())
state_df = pd.melt(state_df.reset_index(), id_vars=['state'])
state_df = state_df.drop('variable',axis=1)
state_df = state_df.set_index(['state'])
```

```python
rep_counts = speaker_df['state'].value_counts().to_frame()
rep_counts = pd.melt(rep_counts.reset_index(), id_vars=['state'])
rep_counts = rep_counts.drop('variable',axis=1)
rep_counts.columns = ['value', 'state']
rep_counts = rep_counts.set_index(['state'])
```

```python
state_words = (state_df.value/rep_counts.value).sort_values(ascending=False).to_frame().reset_index()
```

```python
plt.rcParams['figure.figsize']=(11,5)
p = sns.barplot(x= state_words.state, y=state_words.value)
p.set_xticklabels(p.get_xticklabels(), rotation=90)
p.set(xlabel='State', ylabel='State Words per Rep')
sns.plt.title('State Words per Rep');
```
